```python
path = "../trained_models/model_save/Single-Task_2022-05-28-10-46-17_14035_tgcn_tgcn_best_model/roc.joblib"
```

```python
roc = joblib.load(path)
roc
```

```python
mean_fpr=np.linspace(0,1,100)
tprs = []
for one_roc in roc:
    tprs.append(np.interp(mean_fpr, one_roc["fpr"], one_roc["tpr"]))
```

```python
mean_tprs = np.mean(tprs, axis=0)
```

```python
for one_roc in roc:
    sns.lineplot(one_roc["fpr"], one_roc["tpr"], alpha=0.1)
sns.lineplot(mean_fpr, mean_tprs, alpha=1)
```
